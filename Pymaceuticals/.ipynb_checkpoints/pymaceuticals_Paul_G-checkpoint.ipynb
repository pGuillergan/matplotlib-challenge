```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = clinical_trial_df.merge(drug_data_df, on='Mouse ID', how='left')

# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_response_sr = combined_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean()
# Convert to DataFrame
tumor_response_df = tumor_response_sr.to_frame().reset_index()
# Preview DataFrame
tumor_response_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem_sr = combined_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
tumor_sem_df = tumor_sem_sr.to_frame().reset_index()
# Preview DataFrame
tumor_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_response_rformat_df = tumor_response_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
tumor_response_rformat_df.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = list(tumor_response_rformat_df.index)

Capomulin, = plt.plot(x_axis, tumor_response_rformat_df[['Capomulin']], 
                      color="red", label="Capomulin", 
                      marker='o', linestyle='dashed' )
Infubinol, = plt.plot(x_axis, tumor_response_rformat_df[['Infubinol']], 
                      color="blue", label="Infubinol", 
                      marker='s', linestyle='dotted' )
Ketapril, = plt.plot(x_axis, tumor_response_rformat_df[['Ketapril']], 
                     color="green", label="Ketapril", 
                     marker='d', linestyle='dashdot' )
Placebo, = plt.plot(x_axis, tumor_response_rformat_df[['Placebo']], 
                    color="purple", label="Placebo", 
                    marker='x', linestyle='solid' )

plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
# Save the Figure
plt.savefig('Tumor Response to Treatment', quality=100, dpi=300 )
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_response_sr = combined_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
meta_response_df = meta_response_sr.to_frame()

# Preview DataFrame
meta_response_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_std_sr = combined_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem()

# Convert to DataFrame
meta_std_df = meta_std_sr.to_frame()

# Preview DataFrame
meta_std_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_response_rf_df = meta_response_sr.to_frame().reset_index()
meta_response_rformat_df = meta_response_rf_df.pivot(index='Timepoint', 
                                                    columns='Drug', 
                                                    values='Metastatic Sites')

# Preview that Reformatting worked
meta_response_rformat_df.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = list(meta_response_rformat_df.index)

Capomulin, = plt.plot(x_axis, meta_response_rformat_df[['Capomulin']], 
                      color="red", label="Capomulin", 
                      marker='o', linestyle='dashed' )
Infubinol, = plt.plot(x_axis, meta_response_rformat_df[['Infubinol']], 
                      color="blue", label="Infubinol", 
                      marker='s', linestyle='dotted' )
Ketapril, = plt.plot(x_axis, meta_response_rformat_df[['Ketapril']], 
                     color="green", label="Ketapril", 
                     marker='d', linestyle='dashdot' )
Placebo, = plt.plot(x_axis, meta_response_rformat_df[['Placebo']], 
                    color="purple", label="Placebo", 
                    marker='x', linestyle='solid' )

plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met Sites")
plt.grid()
# Save the Figure
plt.savefig('Metastatic Spread During Treatment', quality=100, dpi=300 )
# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rates_sr = combined_data.groupby(['Drug', 'Timepoint'])['Mouse ID'].count()
# Convert to DataFrame
survival_rates_df = survival_rates_sr.to_frame().reset_index()
# Preview DataFrame
survival_rates_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_rates_rformat_df = survival_rates_df.pivot(index='Timepoint', 
                                                    columns='Drug', 
                                                    values='Mouse ID')
# Preview the Data Frame
survival_rates_rformat_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
x_axis = list(survival_rates_rformat_df.index)

Capomulin, = plt.plot(x_axis, survival_rates_rformat_df[['Capomulin']], 
                      color="red", label="Capomulin", 
                      marker='o', linestyle='dashed' )
Infubinol, = plt.plot(x_axis, survival_rates_rformat_df[['Infubinol']], 
                      color="blue", label="Infubinol", 
                      marker='s', linestyle='dotted' )
Ketapril, = plt.plot(x_axis, survival_rates_rformat_df[['Ketapril']], 
                     color="green", label="Ketapril", 
                     marker='d', linestyle='dashdot' )
Placebo, = plt.plot(x_axis, survival_rates_rformat_df[['Placebo']], 
                    color="purple", label="Placebo", 
                    marker='x', linestyle='solid' )

plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid()
# Save the Figure
plt.savefig('Survival During Treatment', quality=100, dpi=300 )

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_pct_change_sr = (tumor_response_rformat_df.pct_change().sum())*100
# Display the data to confirm
tumor_pct_change_sr
```

```python
# Store all Relevant Percent Changes into a Tuple
drug_tup = tuple(tumor_pct_change_sr.index)
pct_change_tup = tuple(tumor_pct_change_sr.values)
color_list = [ "green" if i > 0 else "red" for i in pct_change_tup ]
tick_loc_y = [ 10 if i > 0 else -10 for i in pct_change_tup ]
# Splice the data between passing and failing drugs

plt.bar(drug_tup, pct_change_tup, align = "center", width=1, edgecolor ="white", color=color_list) 

# Orient widths. Add labels, tick marks, etc. 
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
tick_loc_x = plt.xticks(rotation=45)
plt.grid(alpha=0.2)

# Use functions to label the percentages of changes
print(len(pct_change_tup))
print(tick_loc_x[0])
print(tick_loc_y)
for i in range(len(pct_change_tup)):
    plt.text(tick_loc_x[0][i]-0.3, tick_loc_y[i], f'{int(pct_change_tup[i])}%', fontsize=10)

# Save the Figure
plt.savefig('Tumor Change Over 45 Day Treatment', quality=100, dpi=300 )

# Show the Figure
#fig.show()
```
